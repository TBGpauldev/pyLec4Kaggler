## 1. Python 언어
&nbsp;Python 언어를 할 줄 안다는 것은 어디가서 "Python은 이것이다!"라고 자신있게 말 할 수 있어야 한다고 생각합니다. 이번 목차의 내용은 Python 언어에 대한 제 개인적인 설명을 기술하였기 때문에 Python 코딩을 하는데 직접적인 도움이 되는 내용은 아닙니다. 따라서 시간이 없다면 "2. Python 자료형", 자료구조부터 시작해 주세요. 

&nbsp;Python은 인터프리터 기반의 컴퓨터 프로그래밍 언어입니다. 따라서, Python 인터프리터를 설치해야만 개발 혹은 분석이 가능합니다. Python 언어를 조금이라도 시작해본 분들은 Python의 특징으로 '쉽고 간결하다.'라는 문구를 아래와 같은 예제와 함께 많이 보실것입니다.

```c
//C Language
#include <stdio.h>
int main(){
    printf("Hello world");
    return 0;
}
```
```python
# Python Language
print("Hello  world")
```
&nbsp;하지만 Python 언어는 결코 쉬운 언어가 아닙니다. Python은 스크립트 언어 이면서 동시에 프로그래밍 언어입니다. 또한, 객체지향, 함수형등의 멀티 프로그래밍 패러다임을 지원하는 언어입니다. 또한 Dynamic typing 언어로써 변수를 선언할 때, 변수 타입을 지정하지 않은채 변수를 선언합니다. 이 모든 특징들이 Python의 쉬움을 설명하기 위해 언급되는 특징들 이지만, 이러한 특징들이 결국 우리들의 발목을 잡습니다.

&nbsp;Python 언어는 모든것을 지원하지만, 그 모든 것을 지원하기 때문에 사실 다른어떤 언어보다 어려운 언어입니다. 특히, 데이터 사이언티스트, 데이터 애널리스트를 꿈꾸는 케글러들은 대용량 데이터를 필연적으로 다루어야만 하는 운명이기에 짧은 프로그래밍 지식으로 문제를 해결해 나갈 수 없다는 것을 명심해야 합니다.


## 2. 변수 선언 
&nbsp;튜토리얼 첫 내용은, 변수 선언과, 자료형, 자료구조등에 대해 다룰 것입니다. 
 
&nbsp;Python 언어는 Dynamic typing 언어입니다. 이 말은 변수를 선언할 때 다른 언어들 처럼 자료형을 지정하지 않고 변수를 선언한다는 뜻입니다. C언어와 Python의 변수 선언을 비교해 보겠습니다. 
```c 
// C Language
int main(){
    int age = 27; 
}
```
```python
# Python Language
age = 27
```
&nbsp;Python 언어가 Dynamic typing 이라고 해서 자료형이 없는 것이 아닙니다. 모든 정수형, 실수형, 문자열에 대한 자료형이 존재하고, 이러한 자료형은 모두 객체로 구현되어 있습니다. 변수를 선언하고 할당하는 그 순간에, Python 인터프리터는 할당되는 데이터가 정수인지 실수인지 문자인지 혹은 다른 자료형(객체)인지 체크한 후에 알맞는 메모리영역을 할당받아 값을 저장합니다. 

&nbsp;아래 코드는 변수를 선언하고, 값을 할당하는 과정 그리고 해당 변수의 type을 체크하는 코드입니다.

````python
name = 'paulracooni'
age, gender   = 27, 'M'
visual_acuity = 0.8
print(f"Variable name's type is {type(name)}")
print(f"Variable age's type is {type(age)}")
print(f"Variable gender's type is {type(gender)}")
print(f"Variable visual_acuity's type is {type(visual_acuity)}")
print("""
name, age, gender 변수에 할당 된 자료형은 각각
'str', 'int', 'str', 'float'임을 알 수 있으며
이 모든 자료형이 모두 <class 'object name'>임을 알 수 있다. 
""")
````

## 3. Python Type
>**“everything is object in python"**<br>
> Pthon의 모든 Type은 객체이며 Reference Type 입니다. Value Type은 지원하지 않습니다.

Python의 Type은 코드상에서 눈에 보이지 않지만 다양한 Type을 지원합니다. 데이터 타입은 아래와 같이 분류할 수 있습니다. 또한 Type들의 분류마다 각각의 특징이 존재합니다.

````python
line = 'My age is 27'
print(f"""
1. Numeric Typs

    특징: 각각의 Type들은 산술 연산이 가능합니다. 
    -> 1 + 0.0 + 1+2j = {1 + 0.0 + 1+2j}
    
    -  {type(1)}        
    -  {type(0.0)}
    -  {type(1+2j)}
    
2. Container

    1) Sequence Type
    특징: 모든 Sequence Type은 인덱싱이 가능합니다. 
    list([1, 2, 3, 4])[1:2] -> {list([1, 2, 3, 4])[1:2]}
    list([1, 2, 3, 4])[-2:] -> {list([1, 2, 3, 4])[-2:]}
    tuple((1, 2, 3, 4))[0]  -> {tuple((1, 2, 3, 4))[0]}
        - {type(list())}      ex) [1, 2, 3, 4]
        - {type(tuple())}     ex) (1, 2, 3, 4)
        
    2) Text Sequence Type
    특징: Text Sequence 에서만 지원하는 특별한 함수들이 존재한다. 
    'My age is 27'.replace('27', '20') -> 
        - {type(str())}       ex) 'We are kaggler!'

    3) Binary Sequence Type
    특징: 잘 안쓰여요...ㅎ(고급)
        - {type(bytes())}     ex) b'This is binary'
        - {type(bytearray())} ex) b'.\xf0\xf1\xf2'
        
    4) Set Type 
    특징: 데이터 중복을 허용하지 않으며 Set 산술 연산자가 있다.(고급)
    {{1, 2, 3, 4}} & {{3, 4}} -> { set([1,2,3,4])  &  set([3, 4]) }
        - {type(set())}       ex) {{1, 2, 3, 4}}
        
    5) Mapping Type
        - {type(dict())}      ex) {{'name':'paulracooni', 'age':25}}
        
3. Etc..
    - Module, Class, Function, Method, Boolean, Null... 

""")
````

## 4. 제어 흐름
프로그래밍 언어에서 핵심 중 하나는 제어 흐름입니다. <br>

1. 조건제어 - if, else, elif
2. 반복제어 - for, while
3. 제어구문 - break, continue, pass

````python
from collections import Container

numbers = [1,2,3,4,5]
if isinstance(numbers, Container):
    print(numbers)
elif 1 in numbers:
    print(f"1 is in {numbers}")
else: 
    print(f"else conditon")
````

````python
numbers = [1,2,3,4,5]
for number in numbers:
    if numbers[-1] != number:
        print(number, end=' ')
    else:
        print(number)
````

````python
# break - Loop 탈출 
# contimue - Next Loop 
# pass - Next line code excute
for number in numbers:
    if number > 3:
        print(); break
    elif numbers[-1] != number:
        pass
        print(number, end=' ')
        continue
        print('im passed')
    else:
        print(number)
````

````python
# range()    
for number in range(5):
    if range(5)[-1] != number:
        print(number, end=' ')
    else:
        print(number)   
````

````python
# enumerate() 
alphabets = "ABCDE"
for idx, alphabet in enumerate(alphabets):
    if alphabets[-1] != alphabet:
        print(f'{idx} {alphabet}', end=' ')
    else:
        print(f'{idx} {alphabet}')
````

````python
# reversed()
for idx, alphabet in enumerate(reversed(alphabets)):
    if alphabets[0] != alphabet:
        print(f'{idx} {alphabet}', end=' ')
    else:
        print(f'{idx} {alphabet}')        
````

````python
# zip 
for number, alphabet in zip(numbers, alphabets):
    if alphabets[-1] != alphabet:
        print(f'{number} {alphabet}', end=' ')
    else:
        print(f'{number} {alphabet}')        
````

````python
# enumerate + zip 
for idx, (number, alphabet) in enumerate(zip(numbers, alphabets)):
    if alphabets[-1] != alphabet:
        print(f'{idx} {number} {alphabet} /', end=' ')
    else:
        print(f'{idx} {number} {alphabet}')    
````

# 5. 함수
&nbsp;코드를 작성하다 보면 중복되는 코드들을 종종 만나게 됩니다. 이때가 바로 함수를 사용할 때 입니다. 즉 반복되는 부분이 있을 경우, **반복적으로 사용되는 가치 있는 부분**을 한 뭉치로 묶어 **입력값을 주었을때 그에 상응하는 출력값을 주는** 식의 함수로 작성하는 것이 현명합니다. 

&nbsp;함수를 사용하는 또 다른 이유는 자신이 만든 프로그램을 함수화하면 프로그램 흐름을 일목요연하게 볼 수 있기 때문입니다. 마치 공장에서 원재료가 여러 공정을 거쳐 하나의 상품이 되는 것처럼 프로그램에서도 입력한 값이 여러 함수를 거치면서 원하는 결괏값을 내는 것을 볼 수 있습니다. 이렇게 되면 프로그램 흐름도 잘 파악할 수 있고 오류가 어디에서 나는지도 바로 알아차릴 수 있다. 특히 머신러닝 프로젝트에서는 머신러닝 파이프라인을 일목요연하게 함수로 정리하여 순차적으로 관리하는 것 또한 중요합니다.

&nbsp;마침 위 반복문 예제에서 똑같은 코드가 객체만 바뀌어서 작성되어 있는것을 볼 수 있습니다. 위 예제를 함수로 만들어 코드를 간소화 시켜 봅니다. 

````python
def print_container(cont_obj, fin_idx=-1):
    
    for number in cont_obj:
        
        if isinstance(number, tuple):
            
            idx, number = number
            print_line = f'{idx} {number}'
            if isinstance(number, tuple):
                
                number1, number2 = number
                print_line = f'{idx} {number1} {number2}'
        else:
    
            print_line = f'{number}'
        
        print(print_line, end=' ')
            
    print()
````

````python
# range() 
for number in range(5):
    if range(5)[-1] != number:
        print(number, end=' ')
    else:
        print(number)   

print_container(range(5))
````

````python
# enumerate()
alphabets = "ABCDE"
for idx, alphabet in enumerate(alphabets):
    if alphabets[-1] != alphabet:
        print(f'{idx} {alphabet}', end=' ')
    else:
        print(f'{idx} {alphabet}')
        
print_container(enumerate(alphabets))
````

````python
# reversed()
for idx, alphabet in enumerate(reversed(alphabets)):
    if alphabets[0] != alphabet:
        print(f'{idx} {alphabet}', end=' ')
    else:
        print(f'{idx} {alphabet}')   
        
print_container(enumerate(reversed(alphabets)))
````

````python
# zip()
numbers = [1,2,3,4,5]
alphabets = "ABCDE"
for number, alphabet in zip(numbers, alphabets):
    if alphabets[-1] != alphabet:
        print(f'{number} {alphabet}', end=' ')
    else:
        print(f'{number} {alphabet}')        

print_container(zip(numbers, alphabets))
````

````python
# enumerate + zip 
for idx, (number, alphabet) in enumerate(zip(numbers, alphabets)):
    if alphabets[-1] != alphabet:
        print(f'{idx} {number} {alphabet} /', end=' ')
    else:
        print(f'{idx} {number} {alphabet} /')    

print_container(enumerate(zip(numbers, alphabets)))
````
